```python
df['Time_of_Homogenization'] = pd.to_numeric(df['Time_of_Homogenization'], errors='coerce')
df['Time_of_Homogenization'] = df['Time_of_Homogenization'].fillna(0)
```

```python
df['Time_of_Sol_tre'] = pd.to_numeric(df['Time_of_Sol_tre'], errors='coerce')
df['Time_of_Sol_tre'] = df['Time_of_Sol_tre'].fillna(0)
```

```python
df['YS'] = pd.to_numeric(df['YS'], errors='coerce')
df['YS'] = df['YS'].fillna(0)
```

```python
df['TS'] = pd.to_numeric(df['TS'], errors='coerce')
df['TS'] = df['TS'].fillna(0)
```

```python
df['EL'] = pd.to_numeric(df['EL'], errors='coerce')
df['EL'] = df['EL'].fillna(0)
```

```python
df.dtypes
```

```python
import keras
from keras import layers
from keras import ops
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X1 = df.drop(['YS','Homogenization', 'Solution_Treatment', 'Aging', 'Casting_Type'], axis=1)
y1 = df['YS']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

Sequential Model


```python
model = keras.Sequential([
    layers.Dense(45, activation="relu", input_shape=(18,)),  # Input layer with 18 features
    layers.Dense(40, activation="relu"),  # Hidden layer 1
    layers.Dense(35, activation="relu"),  # Hidden layer 2
    layers.Dense(30, activation="relu"),  # Hidden layer 3
    layers.Dense(25, activation="relu"),  # Hidden layer 4
    layers.Dense(20, activation="relu"),  # Hidden layer 5
    layers.Dense(1, activation="sigmoid")  # Output layer with 3 binary outputs
])
```

```python
model.summary()
```

```python
model.compile(
    loss='mse', # mean Squared Error for rgression
    optimizer=keras.optimizers.RMSprop(learning_rate=0.00005),
    metrics=['mae'] # mean Absolute Eror
)
```

```python
history = model.fit(X_train, y_train, batch_size=64, epochs=50, validation_split=0.2)
```

```python
from sklearn.metrics import mean_absolute_error
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', label='Actual')
plt.xlabel("Actual YS")
plt.ylabel("Predicted YS")
plt.title("Actual vs. Predicted YS")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Predicted')
plt.legend()
plt.show()
```

Improvement

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```
